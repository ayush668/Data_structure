```python
4.#Compare and contrast tuples and lists with examples
#Ans
'''
Comparison and Contrast of Tuples and Lists


List: A mutable, ordered collection of items.
Tuple: An immutable, ordered collection of items.
Syntax:

List: Defined using square brackets [].
Tuple: Defined using parentheses ().
Mutability:

List: Mutable - can be modified by adding, removing, or changing elements.
Tuple: Immutable - cannot be modified once created.
Methods:

List: Has many methods like append(), insert(), remove(), pop(), etc.
Tuple: Has limited methods like count() and index(), as tuples cannot be modified.
Performance:

List: Slower due to dynamic resizing and extra memory allocation for modification.
Tuple: Faster and more memory-efficient due to its immutability.
Memory Usage:

List: Requires more memory as extra space is allocated for future additions.
Tuple: Requires less memory due to its fixed nature.
Hashable:

List: Not hashable; cannot be used as dictionary keys or stored in sets.
Tuple: Hashable (if containing only hashable elements), making it suitable as a dictionary key or in sets.
Use Cases:

List: Best suited for collections that need to change (e.g., a list of tasks).
Tuple: Ideal for collections that should remain constant (e.g., coordinates or days of the week).
Error-Prone:

List: More prone to accidental modification, which may lead to bugs.
Tuple: Safer from unintended changes due to immutability.
Nested Structures:

List: Can contain other lists, tuples, or any data types.
Tuple: Can also contain nested lists, tuples, or other data types but is itself immutable.
'''
#Example:
 #list Example:
# List can be modified
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # Adds an element to the list
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'orange']

#Tuple Example:

# Tuple cannot be modified
days = ("Monday", "Tuesday", "Wednesday")
# days[0] = "Sunday"  # This would raise a TypeError
days # Output: ('Monday', 'Tuesday', 'Wednesday')
```

```python
#5. Describe the key features of sets and provide examples of their use.

#Ans.
 #Unordered Collection:

'''Sets are unordered, meaning elements do not have a specific position and may appear in a different order each time.'''

fruits = {"apple", "banana", "cherry"}
print(fruits)  # Output: {'banana', 'apple', 'cherry'} (Order may vary)

#Unique Elements:

'''Sets automatically remove duplicate elements. Only unique values are stored.'''

numbers = {1, 2, 3, 2, 3, 4}
print(numbers)  # Output: {1, 2, 3, 4} (Duplicates removed)

#Mutable:
'''
Sets are mutable, so elements can be added or removed, but the elements themselves must be immutable (e.g., numbers, strings, tuples).
'''

colors = {"red", "green", "blue"}
colors.add("yellow")  # Adding an element
print(colors)  # Output: {'red', 'green', 'blue', 'yellow'}

colors.remove("green")  # Removing an element
print(colors)  # Output: {'red', 'blue', 'yellow'}

#Unindexed:
'''
Since sets are unordered, they do not support indexing, slicing, or other sequence-like behavior.
'''

fruits = {"apple", "banana", "cherry"}
# fruits[0]  # This will raise a TypeError, as sets are unindexed

#Efficient Operations:

'''Sets provide efficient operations for testing membership (in), adding elements, and set operations (union, intersection, difference).
'''
fruits = {"apple", "banana", "cherry"}
print("apple" in fruits)  # Output: True (Membership test)
#Set Operations:
'''
Sets support mathematical set operations like union, intersection, difference, and symmetric difference.
'''
#Union: Combines elements from both sets, removing duplicates.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2))  # Output: {1, 2, 3, 4, 5}

#Intersection: 
'''Returns elements that are common to both sets.'''

print(set1.intersection(set2))  # Output: {3}
#Difference:
'''Returns elements present in the first set but not in the second.
'''
print(set1.difference(set2))  # Output: {1, 2}
#Symmetric Difference: 
'''Returns elements that are in either of the sets, but not in both.
'''

print(set1.symmetric_difference(set2))  # Output: {1, 2, 4, 5}
```

```python
#Discuss the use cases of tuples and sets in Python programming
'''
Tuples and sets are both important data structures in Python, each with distinct characteristics and use cases. Here's an overview of their use cases:
'''
#Tuples:
#Immutable Collections:
'''
Tuples are immutable, meaning once they are created, their contents cannot be changed. This makes them suitable for storing data that should not be modified, providing a level of data integrity.
Use case: Storing constant values, like coordinates (x, y) or RGB values (255, 0, 0), where the data should remain unchanged throughout the program.
'''

#Heterogeneous Data:
'''
Tuples can store elements of different data types. This is useful when you need to bundle related data together.
Use case: Storing a combination of related values, such as an employee’s ID, name, and salary in a tuple.
'''
#As Dictionary Keys:
'''
Due to their immutability, tuples can be used as keys in dictionaries, while lists (which are mutable) cannot.
Use case: Storing complex keys (like (latitude, longitude)) in a dictionary to map geolocation data to specific values.
'''
#Function Return Values:
'''
Tuples can be used to return multiple values from a function.
Use case: When a function needs to return multiple results, such as divmod(a, b) returning both quotient and remainder, or a function returning both a status and an error message.
'''
#Efficient Memory Usage:
'''
Tuples are more memory-efficient than lists because of their immutability. This makes them ideal for storing large collections of data when modification is not required.
Use case: Storing large datasets that need to be accessed but not altered (e.g., a fixed configuration or lookup data).
'''
#Sets:
#Unique Elements:
'''
Sets automatically remove duplicate entries, making them ideal when you need to store only unique values.
Use case: Tracking unique items, such as unique customer IDs or unique words in a text.
'''
#Efficient Membership Testing:
'''
Sets offer average O(1) time complexity for membership tests (i.e., checking if an element exists in the set), making them faster than lists for such operations.
Use case: Checking if an element exists in a collection, like determining if a certain product ID is already in a set of purchased items.
'''
#Set Operations (Union, Intersection, Difference):

'''Sets support mathematical operations like union, intersection, difference, and symmetric difference, which can be very useful when dealing with sets of data.
Use case: Finding common elements between two lists (using intersection), or identifying what’s in one set but not another (using difference).
'''
#Removing Duplicates:
'''
Sets are an excellent way to remove duplicates from a list or other iterable since they automatically ensure that only unique elements remain.
Use case: Removing duplicate values from a list, such as a list of email addresses, where you want to ensure no duplicate entries are stored.
'''
#Fast Data Processing:
'''
Sets are often used when the order of elements does not matter, and fast operations like adding, removing, and checking membership are needed.
Use case: Analyzing large datasets to find unique items, such as filtering out unique words from a collection of documents.

'''
```

```python
#Describe how to add, modify, and delete items in a dictionary with examples
#1. Adding Items to a Dictionary:
#You can add items to a dictionary by assigning a new key with a value.

#Example:

# Creating a dictionary
my_dict = {'name': 'Alice', 'age': 25}

# Adding a new item
my_dict['city'] = 'New York'

print(my_dict)
# Output: {'name': 'Alice', 'age': 25, 'city': 'New York'} #In this example, a new key 'city' is added to the dictionary with the value 'New York'.

#2. Modifying Items in a Dictionary:
#You can modify an existing value by referencing the key and assigning a new value to it.

#Example:

# Modifying an existing item
my_dict['age'] = 26

print(my_dict)
# Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}
#Here, the value associated with the key 'age' is updated from 25 to 26.

#3. Deleting Items from a Dictionary:
#You can delete items using the del statement or the .pop() method.

#Using del:
#The del keyword removes an item from the dictionary based on its key.

#Example:

# Deleting an item using 'del'
del my_dict['city']

print(my_dict)
# Output: {'name': 'Alice', 'age': 26}
#In this example, the key-value pair 'city': 'New York' is deleted from the dictionary.

#Using .pop():
#The .pop() method removes and returns the value for a given key. If the key does not exist, it raises a KeyError, unless a default value is provided.

#Example:

# Deleting an item using 'pop'
age_value = my_dict.pop('age')

print(my_dict)
# Output: {'name': 'Alice'}
print("Removed value:", age_value)
# Output: Removed value: 26
#In this case, the key 'age' is removed, and its value 26 is returned and printed.

#4. Removing All Items from a Dictionary:
#To remove all items from a dictionary, you can use the .clear() method.

#Example:

# Clearing the dictionary
my_dict.clear()

print(my_dict) # Output: {}
```

```python
#8.Discuss the importance of dictionary keys being immutable and provide example.

#1. Using Numbers as Dictionary Keys:

#Numbers, such as integers and floats, are immutable types in Python. This makes them suitable for use as dictionary keys.

# Example:

# Using integers as dictionary keys
my_dict = {1: "apple", 2: "banana"}
print(my_dict)
# Output: {1: 'apple', 2: 'banana'}

# Using floats as dictionary keys
my_dict = {3.14: "pi", 2.71: "e"}
print(my_dict)
# Output: {3.14: 'pi', 2.71: 'e'}


#In this example, both integers and floats are valid dictionary keys because they are immutable.


#2. Using Strings as Dictionary Keys:

#Strings are another immutable type in Python, making them a common choice for dictionary keys.

#Example:

# Using strings as dictionary keys
my_dict = {"name": "Alice", "age": 25}
print(my_dict)
# Output: {'name': 'Alice', 'age': 25}
#Here, the dictionary uses strings (`"name"` and `"age"`) as keys, which is perfectly valid because strings are immutable.
#3. Using Tuples as Dictionary Keys:

#Tuples are immutable, which means they can be used as dictionary keys, as long as all the elements inside the tuple are also immutable.

#Example:

# Using tuples as dictionary keys
my_dict = {("x", "y"): 10, ("a", "b"): 20}
print(my_dict)
# Output: {('x', 'y'): 10, ('a', 'b'): 20}


#In this example, the tuples ("x", "y") and ("a", "b") are used as dictionary keys because they are immutable.
```
